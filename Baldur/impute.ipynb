# Preliminaries

```python
# importing packages
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# importing the challenge data
import os
origin_dir = os.getcwd()
data_dir = "c:\\Users\\chin.kwan\\OneDrive - IFRC\\Desktop\\Missing data imputation\\fdrs_imputation_2.0\\challenge_data"
os.chdir(data_dir)
train_data_files = os.listdir(data_dir)[:-3]
train_data = pd.DataFrame()

for file in train_data_files:
    chal_data = pd.read_csv(file)
    train_data = pd.concat([train_data, chal_data])
    
# Adding ISO3 code
ns_data = pd.read_csv("ns_data.csv")
train_data = train_data.merge(ns_data[["KPI_DON_code", "iso_3"]], how = "left", left_on = "doncode", right_on = "KPI_DON_code")
```

```python
# Changing back directory
os.chdir(origin_dir)

# Transforming data into wide format
train_data = train_data.drop("doncode", axis = 1).pivot(index = ["iso_3", "year"], columns = "kpi", values = "value").reset_index()
train_data.columns.name = ""
train_data.head()
```
